```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import pandas as pd
import io
import requests
cases_url="https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"
deaths_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv"
cases_content=requests.get(cases_url).content
deaths_content=requests.get(deaths_url).content
corona_cases=pd.read_csv(io.StringIO(cases_content.decode('utf-8')))
corona_deaths=pd.read_csv(io.StringIO(deaths_content.decode('utf-8')))

us_cases = corona_cases.loc[corona_cases['Country/Region'] == 'US']
us_cases.drop(['Province/State', 'Country/Region', 'Lat', 'Long'], axis=1, inplace = True)
us_cases = us_cases.transpose()
us_cases.name = 'US'
us_cases.columns = ['cases']

cases_today = us_cases[-1:]['cases'][0]
cases_yesterday = us_cases[-2:]['cases'][0]
print("STATISTICS FOR THE UNITED STATES\n")
print("Cases yesterday: ", cases_yesterday)
print("Cases today: ", cases_today)
print("New cases today: ", cases_today - cases_yesterday)
print("Percent Change: {0}%".format(round((cases_today - cases_yesterday) / cases_yesterday * 100)))
print("------------------------------------------------------")
us_cases.plot(style='.-')

us_deaths = corona_deaths.loc[corona_deaths['Country/Region'] == 'US']
us_deaths.drop(['Province/State', 'Country/Region', 'Lat', 'Long'], axis=1, inplace = True)
us_deaths = us_deaths.transpose()
us_deaths.columns = ['deaths']

deaths_today = us_deaths[-1:]['deaths'][0]
deaths_yesterday = us_deaths[-2:]['deaths'][0]

print("Deaths yesterday: ", deaths_yesterday)
print("Deaths today: ", deaths_today)
print("New deaths today: ", deaths_today - deaths_yesterday)
print("Percent Change: {0}%".format(round((deaths_today - deaths_yesterday) / deaths_yesterday * 100)))

us_deaths.plot(style='.-')
```

```python
india_cases = corona_cases.loc[corona_cases['Country/Region'] == 'India']
india_cases.head()
india_cases.drop(['Province/State', 'Country/Region', 'Lat', 'Long'], axis=1, inplace = True)
india_cases = india_cases.transpose()
india_cases.columns = ['cases']
india_cases.name = 'India'

cases_today = india_cases[-1:]['cases'][0]
cases_yesterday = india_cases[-2:]['cases'][0]
print("STATISTICS FOR INDIA\n")
print("Cases yesterday: ", cases_yesterday)
print("Cases today: ", cases_today)
print("New cases today: ", cases_today - cases_yesterday)
print("Percent Change: {0}%".format(round((cases_today - cases_yesterday) / cases_yesterday * 100)))
print("------------------------------------------------------")
india_cases.plot(style='.-')

india_deaths = corona_deaths.loc[corona_deaths['Country/Region'] == 'India']
india_deaths.drop(['Province/State', 'Country/Region', 'Lat', 'Long'], axis=1, inplace = True)
india_deaths = india_deaths.transpose()
india_deaths.columns = ['deaths']

deaths_today = india_deaths[-1:]['deaths'][0]
deaths_yesterday = india_deaths[-2:]['deaths'][0]

print("Deaths yesterday: ", deaths_yesterday)
print("Deaths today: ", deaths_today)
print("New deaths today: ", deaths_today - deaths_yesterday)
print("Percent Change: {0}%".format(round((deaths_today - deaths_yesterday) / deaths_yesterday * 100)))

india_deaths.plot(style='.-')
```

```python
fig = plt.figure()
plt.figure(figsize=(18,6))
plt.xticks(rotation='vertical')

for frame in [us_cases, italy_cases, india_cases]:
    plt.plot(frame['cases'], label=frame.name)
    
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
```
